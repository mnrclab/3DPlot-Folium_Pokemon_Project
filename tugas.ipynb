# **Data Visualization - 3D Plot**

Sumber data: https://www.bps.go.id/statictable/2009/02/20/1267/penduduk-indonesia-menurut-provinsi-1971-1980-1990-1995-2000-dan-2010.html

Instruksi tugas
- Dari data, buatlah Plot menggunakan barplot dalam 3d projection
- Data di-plot per tahun

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as im
import matplotlib.cm as cm
import seaborn as sb
import mpl_toolkits.mplot3d.axes3d
```

```python
df = pd.read_excel(
    'indo.xlsx',
    index_col='Unnamed: 0',
    na_values=['-'],
    header = 3,
    skipfooter = 3
)
df
```

```python
list_of_color = ['r','g','b','y','k','r','g','b','y','k','r','g','b','y','k','r','g','b','y','k','r','g','b','y','k','r','g','b','y','k','r','g','b']
df = df.fillna(0)
df
```

### **Penduduk Indonesia 1971**

```python
fig = plt.figure('Tes 3D', figsize=(15,8))
p = plt.subplot(111, projection='3d')

x = np.arange(len(df.index))
y = np.arange(len(df.index)) #data y
z = np.zeros(len(df.index)) #titik mulainya bar

dx = np.ones(len(df.index))
dy = np.ones(len(df.index))
dz = df[1971] #data z

p.bar3d(x, y, z, dx,dy,dz, 
        color= list_of_color)
p.set_ylabel('Tahun 1971')
p.set_zlabel('Populasi 1971')
p.set_xticks(np.arange(len(df.index)))
p.set_xticklabels(df.index, rotation=90)
plt.show()
```

### **Penduduk Indonesia 1980**

```python
fig = plt.figure('Tes 3D', figsize=(15,8))
p = plt.subplot(111, projection='3d')

x = np.arange(len(df.index))
y = np.arange(len(df.index)) #data y
z = np.zeros(len(df.index)) #titik mulainya bar

dx = np.ones(len(df.index))
dy = np.ones(len(df.index))
dz = df[1980] #data z

p.bar3d(x, y, z, dx,dy,dz,
       color = cm.rainbow(np.linspace(0, 1, len(dz)))
       )
p.set_ylabel('Tahun 1980')
p.set_zlabel('Populasi 1980')
p.set_xticks(np.arange(len(df.index)))
p.set_xticklabels(df.index, rotation=90)
plt.show()
```
