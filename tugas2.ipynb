## **Visualization Using Folium**

```python
import folium
import pandas as pd
```

### **A. Akses Data**

```python
df = pd.read_csv('pokemon.csv')
df = df.head(1000)

lat = df['lat']
lng = df['lng']
num = df['num']
name = df['name']
df.head(5)
```

### **B. Creating Folium Map (Standar Icon & Radius)**

```python
map = folium.Map(
    location = [lat[0],lng[0]],
    tiles = 'OpenStreetMap',
    zoom_start = 12)
    
for i in range(len(num)):
    folium.Marker(
        location = [lat[i],lng[i]],
        tooltip = f'<img width="100px" src="https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/back/{num[i]}.png">',
        icon = folium.Icon(color='pink', prefix='fa', icon='user')
    ).add_to(map)

    folium.Circle(
        location = [lat[i],lng[i]],
        radius = 200,
        color = 'red', fill = True, fill_color='yellow'
    ).add_to(map)

map
```

### **C. Creating Folium With Custom Icon**

```python
map = folium.Map(
    location = [lat[0],lng[0]],
    tiles = 'OpenStreetMap',
    zoom_start = 12)
    
for i in range(len(num)):
    icon_url = f'https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/back/{num[i]}.png'
    icon = folium.features.CustomIcon(icon_url,icon_size=(28, 30))
    folium.Marker(
        location = [lat[i],lng[i]],
        tooltip = f"<img width='100px' src='https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/{num[i]}.png'><br><b><center>{name[i]}</center></b>",
        icon = icon
    ).add_to(map)

map
```
